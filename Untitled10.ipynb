```python
# Clean and convert Amount column while preserving sign
if df is not None:
    print("Cleaning and converting Amount column:")

    # Function to clean amount values
    def clean_amount(value):
        if pd.isna(value):
            return value
        if isinstance(value, str):
            # Remove non-breaking spaces and strip whitespace
            cleaned = value.replace('\xa0', '').strip()
            # Convert to float, preserving negative sign
            try:
                return float(cleaned)
            except ValueError:
                return np.nan
        return value

    # Apply cleaning function
    df['Amount'] = df['Amount'].apply(clean_amount)

    print("After cleaning - Amount column info:")
    print(f"Data type: {df['Amount'].dtype}")
    print(f"Null values: {df['Amount'].isnull().sum()}")
    print(f"Non-null values: {df['Amount'].notnull().sum()}")

    # Verify the cleaned values
    print("\nSample of cleaned Amount values (including previously problematic ones):")
    sample_indices = list(range(5)) + [2361, 2362, 2363, 2364, 2366, 2371]
    for idx in sample_indices:
        if idx < len(df):
            print(f"Index {idx}: {df.loc[idx, 'Amount']}")
```

```python
# Implement intelligent time imputation for missing Time values
if df is not None:
    print("INTELLIGENT TIME IMPUTATION")
    print("=" * 50)

    # First, let's identify and analyze the rows with missing Time
    missing_time_rows = df[df['Time'].isnull()].copy()
    print(f"Rows with missing Time: {len(missing_time_rows)}")

    # Show the details of these rows
    print("\nRows needing time imputation:")
    print(missing_time_rows[['Country', 'Indicator', 'Frequency', 'Amount']].to_string())

    # Create a copy to work with for imputation
    df_imputed = df.copy()

    # Get indices of rows with missing Time
    missing_time_mask = df_imputed['Time'].isnull()
    missing_indices = df_imputed[missing_time_mask].index

    print(f"\nStarting imputation for {len(missing_indices)} rows...")

    # Impute Time based on Frequency and country patterns
    for idx in missing_indices:
        country = df_imputed.loc[idx, 'Country']
        frequency = df_imputed.loc[idx, 'Frequency']

        # Get other valid data from same country and frequency
        similar_data = df_imputed[
            (df_imputed['Country'] == country) &
            (df_imputed['Frequency'] == frequency) &
            (df_imputed['Time'].notnull())
        ]

        if len(similar_data) > 0:
            # Use the most recent date and add appropriate time period
            most_recent = similar_data['Time'].max()

            if frequency == 'Monthly':
                imputed_time = most_recent + pd.DateOffset(months=1)
            elif frequency == 'Quarterly':
                imputed_time = most_recent + pd.DateOffset(months=3)
            elif frequency == 'Yearly':
                imputed_time = most_recent + pd.DateOffset(years=1)
            else:
                imputed_time = most_recent + pd.DateOffset(months=1)  # default

            df_imputed.loc[idx, 'Time'] = imputed_time
            print(f"Row {idx}: Imputed {imputed_time.date()} for {country} ({frequency})")

    # Update main dataframe and recreate time components
    df = df_imputed
    df['Year'] = df['Time'].dt.year
    df['Month'] = df['Time'].dt.month
    df['Quarter'] = df['Time'].dt.quarter

    print(f"\nImputation complete. Dataset now has {len(df)} rows.")
    print(f"Missing Time values: {df['Time'].isnull().sum()}")
```

```python
# Handle the remaining 59 missing Amount values with intelligent imputation
if df is not None:
    print("INTELLIGENT AMOUNT IMPUTATION")
    print("=" * 50)

    print(f"Missing Amount values before imputation: {df['Amount'].isnull().sum()}")

    # Analyze the pattern of missing Amount values
    missing_amount_info = df[df['Amount'].isnull()]
    print("\nAnalysis of missing Amount values:")
    print("By Country:")
    print(missing_amount_info['Country'].value_counts())
    print("\nBy Year:")
    print(missing_amount_info['Year'].value_counts())
    print("\nBy Frequency:")
    print(missing_amount_info['Frequency'].value_counts())

    # Strategy: Use forward fill within each country and indicator combination
    print("\nImputation Strategy:")
    print("Using forward fill within each Country + Indicator + Frequency group")
    print("This preserves the temporal pattern for each economic indicator")

    # Create a copy for imputation
    df_imputed = df.copy()

    # Group by Country, Indicator, and Frequency, then forward fill within each group
    groups = ['Country', 'Indicator', 'Frequency']
    df_imputed['Amount'] = df_imputed.groupby(groups)['Amount'].transform(
        lambda x: x.ffill().bfill()  # Forward fill, then backward fill if needed
    )

    # For any remaining missing values, use the median by Country and Indicator
    still_missing = df_imputed['Amount'].isnull().sum()
    if still_missing > 0:
        print(f"\n{still_missing} values still missing after forward fill")
        print("Using median imputation by Country and Indicator...")

        # Calculate median by Country and Indicator
        median_by_group = df_imputed.groupby(['Country', 'Indicator'])['Amount'].transform('median')
        df_imputed['Amount'] = df_imputed['Amount'].fillna(median_by_group)

    # Update the main dataframe
    df = df_imputed

    print(f"\nMissing Amount values after imputation: {df['Amount'].isnull().sum()}")

    # Verify the imputation
    print("\nImputation verification:")
    print(f"Total dataset completeness: {df.notnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100:.2f}%")
    print(f"Dataset now has: {len(df)} complete rows")

    # Show sample of imputed values
    print("\nSample of previously missing values (now imputed):")
    previously_missing_indices = missing_amount_info.index
    sample_indices = previously_missing_indices[:5]  # Show first 5
    for idx in sample_indices:
        if idx in df.index:
            print(f"Row {idx}: {df.loc[idx, 'Country']} - {df.loc[idx, 'Indicator']} = {df.loc[idx, 'Amount']:.2f}")
```

```python
# Explain the Amount imputation strategy in detail
if df is not None:
    print("AMOUNT IMPUTATION STRATEGY EXPLANATION")
    print("=" * 50)

    print("WHY THIS STRATEGY WAS CHOSEN:")
    print("1. Temporal Patterns: Economic data has strong time dependencies")
    print("2. Country Context: Each country has unique economic patterns")
    print("3. Indicator Specificity: Different indicators behave differently")
    print("4. Frequency Matters: Monthly, quarterly, yearly data have different patterns")

    print("\n" + "=" * 50)
    print("STEP-BY-STEP IMPUTATION PROCESS:")

    # Show detailed example for one country
    example_country = "Egypt"
    example_indicator = "Budget Deficit/Surplus"

    print(f"\nEXAMPLE: {example_country} - {example_indicator}")
    print("-" * 40)

    # Get the data for this specific group
    egypy_data = df[
        (df['Country'] == example_country) &
        (df['Indicator'] == example_indicator)
    ].sort_values('Time')

    # Find the rows that were originally missing
    originally_missing = [9, 56, 57, 58, 65]  # From our sample output

    print("How missing values were filled:")
    print("Using forward fill within the same Country + Indicator + Frequency group")

    for idx in originally_missing[:3]:  # Show first 3 examples
        if idx in egypy_data.index:
            current_row = egypy_data.loc[idx]
            # Find the previous non-missing value
            previous_data = egypy_data[egypy_data.index < idx]
            if len(previous_data) > 0:
                previous_value = previous_data['Amount'].iloc[-1]
                previous_time = previous_data['Time'].iloc[-1]
                print(f"\nRow {idx} (Time: {current_row['Time'].date()}):")
                print(f"  Was missing, filled with value from {previous_time.date()}: {previous_value:,.2f}")

    print("\n" + "=" * 50)
    print("ALTERNATIVE STRATEGIES CONSIDERED AND REJECTED:")
    print("1. Simple mean/median: Would ignore temporal patterns")
    print("2. Global imputation: Would ignore country-specific contexts")
    print("3. Random imputation: Would introduce unrealistic variability")
    print("4. Dropping rows: Would lose valuable time series continuity")

    print("\n" + "=" * 50)
    print("WHY FORWARD FILL IS APPROPRIATE FOR ECONOMIC DATA:")
    print(" Preserves temporal autocorrelation")
    print(" Maintains country-specific economic trends")
    print(" Respects indicator-specific patterns")
    print(" Keeps the time series continuous")
    print(" Realistic for short-term missing values in economic time series")

    print("\nVALIDATION:")
    print(" All 59 missing values successfully imputed")
    print(" Dataset now 100% complete")
    print(" Temporal patterns preserved")
    print(" Ready for robust time series analysis")
```

```python
# SIMPLE INTERACTIVE DASHBOARD THAT WORKS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets

print("üöÄ Creating African Economic Dashboard...")

def create_simple_dashboard(df):
    """Create a simple but effective interactive dashboard"""

    # Header
    display(HTML("""
    <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 20px; border-radius: 15px; margin-bottom: 20px;">
        <h1 style="color: white; text-align: center; margin: 0;">üåç African Economic Dashboard</h1>
        <p style="color: white; text-align: center; margin: 5px 0 0 0;">Data-Backed Insights & Recommendations</p>
    </div>
    """))

    # Create controls
    countries = sorted(df['Country'].unique())
    indicators = sorted(df['Indicator'].unique())

    country_dropdown = widgets.Dropdown(
        options=countries,
        value=countries[0],
        description='Select Country:',
        style={'description_width': 'initial'}
    )

    analysis_type = widgets.Dropdown(
        options=['GDP Growth Analysis', 'Inflation Analysis', 'Fiscal Health', 'Comprehensive Overview'],
        value='Comprehensive Overview',
        description='Analysis Type:',
        style={'description_width': 'initial'}
    )

    # Output area
    output = widgets.Output()

    def update_dashboard(country, analysis):
        with output:
            clear_output(wait=True)

            if analysis == 'Comprehensive Overview':
                show_comprehensive_overview(country)
            elif analysis == 'GDP Growth Analysis':
                show_growth_analysis(country)
            elif analysis == 'Inflation Analysis':
                show_inflation_analysis(country)
            elif analysis == 'Fiscal Health':
                show_fiscal_analysis(country)

    # Initial display
    display(HTML("<h3>üîç Select Analysis Parameters:</h3>"))
    display(country_dropdown)
    display(analysis_type)
    display(output)

    # Set up interaction
    widgets.interactive(update_dashboard, country=country_dropdown, analysis=analysis_type)

    # Show initial view
    update_dashboard(countries[0], 'Comprehensive Overview')

def show_comprehensive_overview(country):
    """Show comprehensive overview for selected country"""

    country_data = df[df['Country'] == country]

    # Calculate key metrics
    gdp_growth = country_data[country_data['Indicator'] == 'GDP Growth Rate']['Amount'].mean()
    inflation = country_data[country_data['Indicator'] == 'Inflation Rate']['Amount'].mean()
    budget = country_data[country_data['Indicator'] == 'Budget Deficit/Surplus']['Amount'].mean()

    # Display metrics
    metrics_html = f"""
    <div style="display: grid; grid-template-columns: repeat(3, 1fr); gap: 15px; margin: 20px 0;">
        <div style="background: #e8f4fd; padding: 15px; border-radius: 10px; border-left: 4px solid #3498db;">
            <h4 style="margin: 0; color: #2c3e50;">üìà Avg GDP Growth</h4>
            <p style="font-size: 20px; font-weight: bold; color: #27ae60; margin: 5px 0;">{gdp_growth:.1f}%</p>
        </div>
        <div style="background: #ffeaa7; padding: 15px; border-radius: 10px; border-left: 4px solid #f39c12;">
            <h4 style="margin: 0; color: #2c3e50;">üå°Ô∏è Avg Inflation</h4>
            <p style="font-size: 20px; font-weight: bold; color: #e74c3c; margin: 5px 0;">{inflation:.1f}%</p>
        </div>
        <div style="background: #fddede; padding: 15px; border-radius: 10px; border-left: 4px solid #e74c3c;">
            <h4 style="margin: 0; color: #2c3e50;">üí∞ Budget Position</h4>
            <p style="font-size: 20px; font-weight: bold; color: #c0392b; margin: 5px 0;">{budget:,.0f}</p>
        </div>
    </div>
    """
    display(HTML(metrics_html))

    # Create visualizations
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle(f'Comprehensive Economic Analysis: {country}', fontsize=16, fontweight='bold')

    # Plot 1: GDP Growth Trend
    gdp_data = country_data[country_data['Indicator'] == 'GDP Growth Rate']
    if len(gdp_data) > 0:
        axes[0,0].plot(gdp_data['Year'], gdp_data['Amount'], marker='o', linewidth=2, color='green')
        axes[0,0].set_title('GDP Growth Trend', fontweight='bold')
        axes[0,0].set_ylabel('Growth Rate (%)')
        axes[0,0].axhline(y=5, color='red', linestyle='--', alpha=0.7, label='Strong Growth')
        axes[0,0].axhline(y=0, color='black', linestyle='-', alpha=0.5)
        axes[0,0].grid(True, alpha=0.3)
        axes[0,0].legend()

    # Plot 2: Inflation Trend
    inflation_data = country_data[country_data['Indicator'] == 'Inflation Rate']
    if len(inflation_data) > 0:
        axes[0,1].plot(inflation_data['Year'], inflation_data['Amount'], marker='o', linewidth=2, color='orange')
        axes[0,1].set_title('Inflation Trend', fontweight='bold')
        axes[0,1].set_ylabel('Inflation Rate (%)')
        axes[0,1].axhline(y=10, color='red', linestyle='--', alpha=0.7, label='High Inflation')
        axes[0,1].axhline(y=5, color='yellow', linestyle='--', alpha=0.7, label='Moderate')
        axes[0,1].grid(True, alpha=0.3)
        axes[0,1].legend()

    # Plot 3: Budget Position
    budget_data = country_data[country_data['Indicator'] == 'Budget Deficit/Surplus']
    if len(budget_data) > 0:
        colors = ['red' if x < 0 else 'green' for x in budget_data['Amount']]
        axes[1,0].bar(budget_data['Year'], budget_data['Amount'], color=colors, alpha=0.7)
        axes[1,0].set_title('Budget Deficit/Surplus', fontweight='bold')
        axes[1,0].set_ylabel('Amount')
        axes[1,0].axhline(y=0, color='black', linestyle='-', alpha=0.5)
        axes[1,0].grid(True, alpha=0.3)

    # Plot 4: Performance Comparison (placeholder)
    axes[1,1].text(0.5, 0.5, 'Performance Metrics\n& Recommendations',
                   ha='center', va='center', transform=axes[1,1].transAxes, fontsize=12)
    axes[1,1].set_title('Key Insights', fontweight='bold')
    axes[1,1].axis('off')

    plt.tight_layout()
    plt.show()

    # Recommendations
    display(HTML(f"""
    <div style="background: #f8f9fa; padding: 15px; border-radius: 10px; margin: 20px 0;">
        <h3 style="color: #2c3e50;">üéØ Recommendations for {country}</h3>
    </div>
    """))

    # Generate recommendations based on data
    recommendations = generate_recommendations(gdp_growth, inflation, budget)
    for rec in recommendations:
        display(HTML(f"<p>‚Ä¢ {rec}</p>"))

def show_growth_analysis(country):
    """Show detailed GDP growth analysis"""

    country_data = df[df['Country'] == country]
    gdp_data = country_data[country_data['Indicator'] == 'GDP Growth Rate']

    if len(gdp_data) == 0:
        display(HTML(f"<p>No GDP growth data available for {country}</p>"))
        return

    # Calculate growth statistics
    avg_growth = gdp_data['Amount'].mean()
    growth_volatility = gdp_data['Amount'].std()
    best_year = gdp_data.loc[gdp_data['Amount'].idxmax()]
    worst_year = gdp_data.loc[gdp_data['Amount'].idxmin()]

    # Display metrics
    metrics_html = f"""
    <div style="display: grid; grid-template-columns: repeat(2, 1fr); gap: 15px; margin: 20px 0;">
        <div style="background: #e8f4fd; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">üìà Average Growth</h4>
            <p style="font-size: 18px; font-weight: bold; color: #27ae60; margin: 5px 0;">{avg_growth:.1f}%</p>
        </div>
        <div style="background: #ffeaa7; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">üé¢ Growth Volatility</h4>
            <p style="font-size: 18px; font-weight: bold; color: #e67e22; margin: 5px 0;">{growth_volatility:.1f}%</p>
        </div>
    </div>
    """
    display(HTML(metrics_html))

    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Growth trend
    ax1.plot(gdp_data['Year'], gdp_data['Amount'], marker='o', linewidth=2, color='green', label='GDP Growth')
    ax1.fill_between(gdp_data['Year'], gdp_data['Amount'], alpha=0.3, color='green')
    ax1.set_title(f'GDP Growth Trend: {country}', fontweight='bold')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Growth Rate (%)')
    ax1.axhline(y=5, color='red', linestyle='--', label='Strong Growth (5%+)')
    ax1.axhline(y=0, color='black', linestyle='-', alpha=0.5)
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # Growth distribution
    ax2.hist(gdp_data['Amount'], bins=10, color='lightblue', edgecolor='black', alpha=0.7)
    ax2.set_title('Growth Rate Distribution', fontweight='bold')
    ax2.set_xlabel('Growth Rate (%)')
    ax2.set_ylabel('Frequency')
    ax2.axvline(x=avg_growth, color='red', linestyle='--', label=f'Average: {avg_growth:.1f}%')
    ax2.grid(True, alpha=0.3)
    ax2.legend()

    plt.tight_layout()
    plt.show()

    # Insights
    display(HTML(f"""
    <div style="background: #e8f4fd; padding: 15px; border-radius: 10px; margin: 20px 0;">
        <h4 style="color: #2c3e50;">üìä Growth Insights for {country}</h4>
        <p>‚Ä¢ <strong>Best Performance:</strong> {best_year['Year']} ({best_year['Amount']:.1f}%)</p>
        <p>‚Ä¢ <strong>Worst Performance:</strong> {worst_year['Year']} ({worst_year['Amount']:.1f}%)</p>
        <p>‚Ä¢ <strong>Stability:</strong> {'Stable' if growth_volatility < 2 else 'Volatile'} growth pattern</p>
        <p>‚Ä¢ <strong>Trend:</strong> {'Growing' if avg_growth > 3 else 'Stagnant' if avg_growth > 0 else 'Declining'} economy</p>
    </div>
    """))

def show_inflation_analysis(country):
    """Show detailed inflation analysis"""

    country_data = df[df['Country'] == country]
    inflation_data = country_data[country_data['Indicator'] == 'Inflation Rate']

    if len(inflation_data) == 0:
        display(HTML(f"<p>No inflation data available for {country}</p>"))
        return

    # Calculate inflation statistics
    avg_inflation = inflation_data['Amount'].mean()
    inflation_volatility = inflation_data['Amount'].std()
    peak_inflation = inflation_data.loc[inflation_data['Amount'].idxmax()]

    # Display metrics
    inflation_status = "Hyperinflation" if avg_inflation > 50 else "High" if avg_inflation > 10 else "Moderate" if avg_inflation > 5 else "Low"
    status_color = "darkred" if avg_inflation > 50 else "red" if avg_inflation > 10 else "orange" if avg_inflation > 5 else "green"

    metrics_html = f"""
    <div style="display: grid; grid-template-columns: repeat(2, 1fr); gap: 15px; margin: 20px 0;">
        <div style="background: #ffeaa7; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">üå°Ô∏è Average Inflation</h4>
            <p style="font-size: 18px; font-weight: bold; color: {status_color}; margin: 5px 0;">{avg_inflation:.1f}%</p>
            <small style="color: #7f8c8d;">{inflation_status}</small>
        </div>
        <div style="background: #fddede; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">‚ö° Inflation Volatility</h4>
            <p style="font-size: 18px; font-weight: bold; color: #e74c3c; margin: 5px 0;">{inflation_volatility:.1f}%</p>
        </div>
    </div>
    """
    display(HTML(metrics_html))

    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Inflation trend
    ax1.plot(inflation_data['Year'], inflation_data['Amount'], marker='o', linewidth=2, color='red', label='Inflation')
    ax1.fill_between(inflation_data['Year'], inflation_data['Amount'], alpha=0.3, color='red')
    ax1.set_title(f'Inflation Trend: {country}', fontweight='bold')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Inflation Rate (%)')
    ax1.axhline(y=10, color='darkred', linestyle='--', label='High Inflation (10%+)')
    ax1.axhline(y=5, color='orange', linestyle='--', label='Moderate (5%+)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()

    # Inflation categories
    categories = {
        'Low (<5%)': len(inflation_data[inflation_data['Amount'] < 5]),
        'Moderate (5-10%)': len(inflation_data[(inflation_data['Amount'] >= 5) & (inflation_data['Amount'] < 10)]),
        'High (10-20%)': len(inflation_data[(inflation_data['Amount'] >= 10) & (inflation_data['Amount'] < 20)]),
        'Very High (20%+)': len(inflation_data[inflation_data['Amount'] >= 20])
    }

    colors = ['green', 'yellow', 'orange', 'red']
    ax2.pie(categories.values(), labels=categories.keys(), colors=colors, autopct='%1.1f%%', startangle=90)
    ax2.set_title('Inflation Severity Distribution', fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Insights
    display(HTML(f"""
    <div style="background: #ffeaa7; padding: 15px; border-radius: 10px; margin: 20px 0;">
        <h4 style="color: #2c3e50;">üí∞ Inflation Insights for {country}</h4>
        <p>‚Ä¢ <strong>Peak Inflation:</strong> {peak_inflation['Year']} ({peak_inflation['Amount']:.1f}%)</p>
        <p>‚Ä¢ <strong>Stability:</strong> {'Stable' if inflation_volatility < 2 else 'Volatile'} prices</p>
        <p>‚Ä¢ <strong>Risk Level:</strong> {'High' if avg_inflation > 10 else 'Medium' if avg_inflation > 5 else 'Low'}</p>
        <p>‚Ä¢ <strong>Policy Need:</strong> {'Crisis management' if avg_inflation > 20 else 'Tightening' if avg_inflation > 10 else 'Monitoring'}</p>
    </div>
    """))

def show_fiscal_analysis(country):
    """Show detailed fiscal health analysis"""

    country_data = df[df['Country'] == country]
    budget_data = country_data[country_data['Indicator'] == 'Budget Deficit/Surplus']

    if len(budget_data) == 0:
        display(HTML(f"<p>No budget data available for {country}</p>"))
        return

    # Calculate fiscal statistics
    avg_budget = budget_data['Amount'].mean()
    budget_trend = budget_data['Amount'].iloc[-1] - budget_data['Amount'].iloc[0] if len(budget_data) > 1 else 0

    # Display metrics
    budget_status = "Surplus" if avg_budget > 0 else "Deficit"
    status_color = "green" if avg_budget > 0 else "red"

    metrics_html = f"""
    <div style="display: grid; grid-template-columns: repeat(2, 1fr); gap: 15px; margin: 20px 0;">
        <div style="background: #e8f4fd; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">üí∞ Average Budget</h4>
            <p style="font-size: 18px; font-weight: bold; color: {status_color}; margin: 5px 0;">{avg_budget:,.0f}</p>
            <small style="color: #7f8c8d;">{budget_status}</small>
        </div>
        <div style="background: #d5f4e6; padding: 15px; border-radius: 10px;">
            <h4 style="margin: 0; color: #2c3e50;">üìà Budget Trend</h4>
            <p style="font-size: 18px; font-weight: bold; color: #27ae60; margin: 5px 0;">{budget_trend:,.0f}</p>
            <small style="color: #7f8c8d;">{'Improving' if budget_trend > 0 else 'Worsening'}</small>
        </div>
    </div>
    """
    display(HTML(metrics_html))

    # Create visualization
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Budget trend
    colors = ['red' if x < 0 else 'green' for x in budget_data['Amount']]
    ax1.bar(budget_data['Year'], budget_data['Amount'], color=colors, alpha=0.7, edgecolor='black')
    ax1.set_title(f'Budget Position: {country}', fontweight='bold')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Deficit/Surplus')
    ax1.axhline(y=0, color='black', linestyle='-', alpha=0.5)
    ax1.grid(True, alpha=0.3)

    # Budget composition (simplified)
    years_with_deficit = len(budget_data[budget_data['Amount'] < 0])
    years_with_surplus = len(budget_data[budget_data['Amount'] > 0])

    sizes = [years_with_deficit, years_with_surplus]
    labels = ['Deficit Years', 'Surplus Years']
    colors_pie = ['red', 'green']

    ax2.pie(sizes, labels=labels, colors=colors_pie, autopct='%1.1f%%', startangle=90)
    ax2.set_title('Budget Performance History', fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Insights
    fiscal_health = "Critical" if avg_budget < -500000 else "Concerning" if avg_budget < -100000 else "Manageable" if avg_budget < 0 else "Strong"

    display(HTML(f"""
    <div style="background: #e8f4fd; padding: 15px; border-radius: 10px; margin: 20px 0;">
        <h4 style="color: #2c3e50;">üèõÔ∏è Fiscal Insights for {country}</h4>
        <p>‚Ä¢ <strong>Fiscal Health:</strong> {fiscal_health}</p>
        <p>‚Ä¢ <strong>Deficit Years:</strong> {years_with_deficit} out of {len(budget_data)} years</p>
        <p>‚Ä¢ <strong>Trend:</strong> {'Improving' if budget_trend > 0 else 'Deteriorating'} fiscal position</p>
        <p>‚Ä¢ <strong>Action Needed:</strong> {'Urgent consolidation' if fiscal_health == 'Critical' else 'Gradual reforms' if fiscal_health == 'Concerning' else 'Maintain course'}</p>
    </div>
    """))

def generate_recommendations(gdp_growth, inflation, budget):
    """Generate data-driven recommendations"""

    recommendations = []

    # Growth recommendations
    if gdp_growth < 0:
        recommendations.append("üö® **URGENT**: Implement economic stimulus package to reverse recession")
    elif gdp_growth < 3:
        recommendations.append("üìà **PRIORITY**: Boost growth through infrastructure investments and business reforms")
    elif gdp_growth > 5:
        recommendations.append("‚úÖ **MAINTAIN**: Continue growth-friendly policies and focus on sustainability")

    # Inflation recommendations
    if inflation > 20:
        recommendations.append("üî• **CRISIS**: Immediate monetary tightening and inflation control measures needed")
    elif inflation > 10:
        recommendations.append("‚ö†Ô∏è **ALERT**: Implement inflation targeting and supply-side reforms")
    elif inflation > 5:
        recommendations.append("üìä **MONITOR**: Maintain current monetary policy with close inflation watch")
    else:
        recommendations.append("üíö **STABLE**: Low inflation environment - focus on growth objectives")

    # Fiscal recommendations
    if budget < -500000:
        recommendations.append("üí∏ **CRITICAL**: Urgent fiscal consolidation through spending rationalization")
    elif budget < -100000:
        recommendations.append("üìâ **CONCERN**: Gradual deficit reduction through revenue enhancement")
    elif budget < 0:
        recommendations.append("‚öñÔ∏è **MANAGE**: Focus on expenditure efficiency and debt management")
    else:
        recommendations.append("üí™ **STRONG**: Build fiscal buffers and invest in productive sectors")

    # Cross-cutting recommendations
    if gdp_growth > 4 and inflation < 8:
        recommendations.append("üåü **OPPORTUNITY**: Ideal conditions for strategic investments and reforms")

    if gdp_growth < 2 and inflation > 15:
        recommendations.append("üéØ **CHALLENGE**: Stagflation risk - need coordinated monetary and fiscal response")

    return recommendations

# Create comparison view for all countries
def create_country_comparison():
    """Create a comparison view across all countries"""

    display(HTML("""
    <div style="background: #2c3e50; padding: 15px; border-radius: 10px; margin: 20px 0;">
        <h2 style="color: white; text-align: center; margin: 0;">üèÜ Country Performance Comparison</h2>
    </div>
    """))

    # Calculate performance scores for all countries
    performance_data = []

    for country in df['Country'].unique():
        country_data = df[df['Country'] == country]

        gdp_growth = country_data[country_data['Indicator'] == 'GDP Growth Rate']['Amount'].mean() or 0
        inflation = country_data[country_data['Indicator'] == 'Inflation Rate']['Amount'].mean() or 0
        budget = country_data[country_data['Indicator'] == 'Budget Deficit/Surplus']['Amount'].mean() or 0

        # Composite score (higher is better)
        score = (gdp_growth * 2) + max(0, 10 - inflation) + max(0, -budget/100000)

        performance_data.append({
            'Country': country,
            'GDP Growth': gdp_growth,
            'Inflation': inflation,
            'Budget Position': budget,
            'Performance Score': score
        })

    perf_df = pd.DataFrame(performance_data).sort_values('Performance Score', ascending=False)

    # Display ranking table
    display(HTML("<h3>üìä Overall Performance Ranking</h3>"))

    ranking_html = """
    <table style="width: 100%; border-collapse: collapse; margin: 15px 0;">
        <tr style="background: #34495e; color: white;">
            <th style="padding: 10px; border: 1px solid #ddd;">Rank</th>
            <th style="padding: 10px; border: 1px solid #ddd;">Country</th>
            <th style="padding: 10px; border: 1px solid #ddd;">GDP Growth</th>
            <th style="padding: 10px; border: 1px solid #ddd;">Inflation</th>
            <th style="padding: 10px; border: 1px solid #ddd;">Budget</th>
            <th style="padding: 10px; border: 1px solid #ddd;">Score</th>
        </tr>
    """

    for i, (_, row) in enumerate(perf_df.iterrows(), 1):
        bg_color = "#e8f4fd" if i % 2 == 0 else "white"
        ranking_html += f"""
        <tr style="background: {bg_color};">
            <td style="padding: 8px; border: 1px solid #ddd;">{i}</td>
            <td style="padding: 8px; border: 1px solid #ddd; font-weight: bold;">{row['Country']}</td>
            <td style="padding: 8px; border: 1px solid #ddd; color: {'green' if row['GDP Growth'] > 3 else 'red'}">{row['GDP Growth']:.1f}%</td>
            <td style="padding: 8px; border: 1px solid #ddd; color: {'red' if row['Inflation'] > 10 else 'orange' if row['Inflation'] > 5 else 'green'}">{row['Inflation']:.1f}%</td>
            <td style="padding: 8px; border: 1px solid #ddd; color: {'green' if row['Budget Position'] > 0 else 'red'}">{row['Budget Position']:,.0f}</td>
            <td style="padding: 8px; border: 1px solid #ddd; font-weight: bold;">{row['Performance Score']:.1f}</td>
        </tr>
        """

    ranking_html += "</table>"
    display(HTML(ranking_html))

    # Create comparison chart
    fig, ax = plt.subplots(figsize=(12, 8))

    y_pos = np.arange(len(perf_df))
    ax.barh(y_pos, perf_df['Performance Score'], color='skyblue', edgecolor='black')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(perf_df['Country'])
    ax.set_xlabel('Performance Score')
    ax.set_title('Country Economic Performance Ranking', fontweight='bold')
    ax.grid(True, alpha=0.3, axis='x')

    # Add value labels
    for i, v in enumerate(perf_df['Performance Score']):
        ax.text(v + 0.1, i, f'{v:.1f}', va='center', fontweight='bold')

    plt.tight_layout()
    plt.show()

# Initialize the dashboard
print("üåç Loading African Economic Dashboard...")
create_simple_dashboard(df)

# Add country comparison
create_country_comparison()

print("\n" + "="*60)
print("‚úÖ DASHBOARD READY!")
print("="*60)
print("üí° How to use:")
print("   ‚Ä¢ Select different countries from the dropdown")
print("   ‚Ä¢ Choose analysis type (Growth, Inflation, Fiscal, Overview)")
print("   ‚Ä¢ View interactive charts and data-backed recommendations")
print("   ‚Ä¢ Compare country performance in the ranking table")
print("="*60)
```
